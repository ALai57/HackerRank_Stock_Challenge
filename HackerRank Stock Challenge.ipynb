```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('http://hr-filepicker.s3.amazonaws.com/sample.txt', delimiter=' ', skiprows=5, header=None)
```

```python
df = df.T
print(df.info())
print(df.head())
```

```python
col_names = df.loc[0,:]
df.columns = col_names

df.drop(0, inplace=True)
df = df.apply(pd.to_numeric)

print(df.head(2))
print('\n')
df.info()
```

```python
plt.figure(figsize=[12,12])
ax = plt.axes()
for col in df:
   df[col].plot(ax=ax)

plt.legend(df.columns)
plt.show()
```

```python
ax = plt.axes()

df['CAL'].plot(ax=ax)
pd.rolling_mean(df.CAL, window=10).plot(ax=ax)
pd.rolling_std(df.CAL, window=10).plot(ax=ax)
plt.legend(['CAL', 'Mean', 'Std'])
plt.show()
```

```python
plt.figure(figsize=[16,16])

windows = [5, 10, 15]


ax = plt.subplot(3, 3, 1 )
for col in df: 
    df[col].plot(ax=ax)
    #pd.rolling_mean(df.CAL, window=w).plot(ax=ax)
    #pd.rolling_std(df.CAL, window=w).plot(ax=ax)
        
plt.legend(df.columns)
plt.title('Raw data')

ax = plt.subplot(3, 3, 4 )
plt.title('Diff')
for col in df: 
    df[col].diff().plot(ax=ax, legend=None)  

for idx, w in enumerate(windows):  
    ax = plt.subplot(3, 3, 3*idx + 2)
    plt.title('Mean Window = %d' % w)
    for col in df: 
        pd.rolling_mean(df[col], window=w).plot(ax=ax, legend=None)


for idx, w in enumerate(windows):  
    for col in df: 
        ax = plt.subplot(3, 3, 3*(idx+1) )
        plt.title('Std Window = %d' % w)
        pd.rolling_std(df[col], window=w).plot(ax=ax, legend=None)
        
plt.show()
```
